```python
import numpy as np
from scipy.integrate import odeint
from math import sin
```

```python
# initial condition
u0 = 0, 1 

t = np.linspace(0, 10)

def system(y, t):
    y0 = y[0]
    y1 = y[1]
    y2 = -7*y0/3 + sin(5*t)
    return y1, y2

y = odeint(system, u0, t)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(t, y)
```

```python
plt.plot(t, y[:, 0])
```

```python
plt.plot(t, y[:, 1])
```
