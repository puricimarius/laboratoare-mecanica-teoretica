## Sarcina II

```python
from sympy import sympify, lambdify
import numpy as np
import matplotlib.pyplot as plt
```

```python
expr = sympify('3 * exp(-x) * sin(x)')
expr
```

```python
f = lambdify(sympify('x'), expr, 'numpy')
```

```python
expr = sympify('exp(x)  * cos(x)')
expr
```

```python
g = lambdify(sympify('x'), expr, 'numpy')
```

```python
x = np.arange(0, 3 + 0.0001, 0.0001)
```

```python
plt.figure(figsize=(12, 8), dpi=300)
plt.title(r'Graficele funcÈ›iilor $f\left(x\right)$ È™i $g\left(x\right)$ pe aceleaÈ™i axe')
plt.plot(x, f(x), color='b')
plt.plot(x, g(x), color='g')
plt.legend([r'$f\left(x\right)$', r'$g\left(x\right)$'])
plt.xlabel(r'$x$ - axis')
plt.ylabel(r'$y$ - axis')
plt.grid()
```

### Graficele funcÈ›iilor Ã®ntr-o fereastrÄƒ, pe axe diferite, orizontal

```python
fig = plt.figure(figsize=(14, 9), dpi=300)
functions = [[(f, 'b', r'$f(x)$')], [(g, 'r', r'$g(x)$')], [(f, 'b', r'$f(x)$'), (g, 'r', r'$g(x)$')]]
for i in range(3):
    ax = fig.add_subplot(311 + i)
    for func in functions[i]:
        ax.plot(x, func[0](x), color=func[1])
    ax.legend([ item[2] for item in functions[i] ])
    ax.grid()
fig.tight_layout()
```

### Graficele funcÈ›iilor Ã®ntr-o fereastrÄƒ, pe axe diferite, vertical

```python
fig = plt.figure(figsize=(9, 6), dpi=300)
for i in range(3):
    ax = fig.add_subplot(131 + i)
    for func in functions[i]:
        ax.plot(x, func[0](x), color=func[1])
    ax.legend([ item[2] for item in functions[i] ])
    ax.grid()
fig.tight_layout()
```

### Graficele funcÈ›iilor Ã®ntr-o fereastrÄƒ, fiecare grafic pe axe diferite È™i apoi ambele Ã®mpreunÄƒ

```python
fig = plt.figure(figsize=(9, 6), dpi=300)
order = (221, 223, 122)
for i in range(3):
    ax = fig.add_subplot(order[i])
    for func in functions[i]:
        ax.plot(x, func[0](x), color=func[1])
    ax.legend([ item[2] for item in functions[i] ])
    ax.grid()
fig.tight_layout()
```

### Graficele funcÈ›iilor Ã®ntr-o fereastrÄƒ, fiecare grafic pe axe diferite È™i apoi ambele Ã®mpreunÄƒ

```python
fig = plt.figure(figsize=(9, 6), dpi=300)
order = (223, 224, 211)
for i in range(3):
    ax = fig.add_subplot(order[i])
    for func in functions[i]:
        ax.plot(x, func[0](x), color=func[1])
    ax.legend([ item[2] for item in functions[i] ])
    ax.grid()
fig.tight_layout()
```

## Sarcina III

```python
expr = sympify('2*x*sin(x + y)*log(Abs(y) + 2)')
expr
```

```python
z = lambdify((sympify('x'), sympify('y')), expr, 'numpy')
x = np.arange(-np.pi, np.pi + 0.001, 0.001)
y = np.arange(-2, 2 + 0.001, 0.001)
x, y = np.meshgrid(x, y)
```

```python
from matplotlib import cm

fig = plt.figure(figsize=(6, 6), dpi=300)
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, z(x, y), cmap=cm.coolwarm)
```

```python
plt.contour(x, y, z(x, y))
```
